# First few steps for learning Python

1. Print Function -> print()

2. Variable

3. Data Type -> type()

4. Comment

5. Mutiable

```python
print("Hello") # String

name = "Adil" # Variable should have a meaningful name

print(f"Hello From {name}")

print(10+2) #
```

```python
# Storing a single line (Single line Comment)
var = "Hello There"

print(var)

# Checking data type
print(type(var)) # string
print(type(2025)) # integer
print(type(2025.01)) # float

"""
This is a
multi line
comment
"""
```

```python
# String is Immutable

name = "Sakib"

#name[0] = "R" # TypeError: 'str' object does not support item assignment
```

```python
# List is mutable
list = [1, 2, 3, "Adil"]

print(list)

list[0] = 5

list.append("Rahman")

print(list)

list.append(6)

print(list)
```

```python
text = "Programer Keno Hoba?"

print(text[0])

print(text[-2])

text[::-1]
```

```python
language = "Python"

language[0:3] # last index is excluded

language[2:5] # Slicing
```

```python
# Concatenation
first_name = "Adil"
last_name = "Rahman"

full_name = first_name+" "+last_name

print(full_name)
```

```python
# multiplecation
10*2
```

```python
# Membership
language = "Python"

print(language*3)

print("Py" in language)
```

```python
# Operations

a = 10
b = 3

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b) # Floor division
print(a%b)
```
